# PROBLEMA 5

Dada la sucesion.

1. **Valores de \( a_1 \) a \( a_5 \)**:
   \[
   \begin{aligned}
       a_1 &\approx 0.8415, \\
       a_2 &\approx 0.9589, \\
       a_3 &\approx 0.9816, \\
       a_4 &\approx 0.9896, \\
       a_5 &\approx 0.9933.
   \end{aligned}
   \]

2. **Valores de \( a_n \) para \( n = 1, 2, 3, \ldots, 200 \)**:
   - Los valores se calcularon y se muestran en el gráfico.

3. **Gráfico de \( a_n \) vs \( n \)**:
   - El gráfico muestra los puntos \((n, a_n)\) para \( n = 1, 2, 3, \ldots, 200 \), junto con la línea horizontal \( y = 1 \).

4. **Conclusión sobre el límite**:
   - Parece razonable que \(\lim_{n \to \infty} a_n = 1\).

5. **Valor más pequeño de \( n \) para el cual \( a_n > 0.999 \)**:
   - \( n = 13 \).

6. **Primer valor de \( n \) para el cual \( |a_n - 1| < 0.0001 \)**:
   - \( n = 41 \).

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function a_n
def a_n(n):
    return n * np.sin(1 / n)

# Compute a_n for n = 1 to 200
n_values = np.arange(1, 201)
a_n_values = a_n(n_values)

# Plot the points (n, a_n) and the horizontal line y = 1
plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n_values, label=r'$a_n = n \sin\left(\frac{1}{n}\right)$', color='blue')
plt.axhline(y=1, color='red', linestyle='--', label='$y = 1$')
plt.xlabel('$n$')
plt.ylabel('$a_n$')
plt.title('Gráfico de $a_n$ vs $n$')
plt.legend()
plt.grid(True)
plt.show()
```

# PROBLEMA 9


Dada la secuencia

1. **Calcular los primeros términos de la secuencia**:
   \[
   \begin{aligned}
       a_1 &= 2, \\
       a_2 &= 2 \times 2 = 4, \\
       a_3 &= 2 \times 4 = 8, \\
       a_4 &= 2 \times 8 = 16, \\
       a_5 &= 2 \times 16 = 32.
   \end{aligned}
   \]

2. **Graficar los puntos \((n, a_n)\) para \(n = 1, 2, 3, \ldots, 50\)**:
   - El gráfico muestra cómo \(a_n\) crece exponencialmente a medida que \(n\) aumenta.

3. **¿Tiene la secuencia un límite?**:
   - No, la secuencia no tiene un límite. Dado que cada término es el doble del anterior, la secuencia diverge a \(\infty\).

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sequence a_n
def a_n(n, a_1=2):
    return a_1 * 2**(n-1)

# Compute a_n for n = 1 to 50
n_values = np.arange(1, 51)
a_n_values = a_n(n_values)

# Plot the points (n, a_n)
plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n_values, marker='o', linestyle='-', color='b', label=r'$a_{n+1} = 2a_n$')
plt.xlabel('$n$')
plt.ylabel('$a_n$')
plt.title('Gráfico de $a_n$ vs $n$')
plt.legend()
plt.grid(True)
plt.yscale('log')  # Use log scale for better visualization of exponential growth
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sequence a_n
def a_n(n, a_1=10):
    return a_1 * (1/3)**(n-1)

# Compute a_n for n = 1 to 50
n_values = np.arange(1, 51)
a_n_values = a_n(n_values)

# Plot the points (n, a_n)
plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n_values, marker='o', linestyle='-', color='b', label=r'$a_{n+1} = \frac{1}{3}a_n$')
plt.xlabel('$n$')
plt.ylabel('$a_n$')
plt.title('Gráfico de $a_n$ vs $n$')
plt.legend()
plt.grid(True)
plt.show()
```
