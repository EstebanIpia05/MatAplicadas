**METODOLOGÍA:**<p>
Tiempo:<p>
$T(x)=\frac{d(x)}{v(x)}$

Donde:
- d(x) es la distancia de la ruta en función de un parámetro x (hora).
- v(x) es la velocidad promedio en dicha ruta.

Ejemplo:

**x = 0 - Ruta 1:**<p>
d(0) = 10 km<p>
v(0) = 20 km/h

$T(0)=\frac{10}{20}=0,5 horas$

**x = 1 - Ruta 2:**<p>
d(1) = 20 km<p>
v(1) = 40 km/h

$T(0)=\frac{20}{40}=0,5 horas$


Es decir, que por las dos rutas se toma el mismo tiempo en llegar de un punto A a un punto B, porque en uno es más corta la distancia pero la velocidad es lenta y en el otro es lo contrario, como podemos evidenciar en este ejemplo.

$T(x)=\frac{d(x)}{v(x)}=\frac{5x^2-10x+50}{-2x+30}$

5x^2: Esto indica que a ciertas horas el desvío o tráfico puede hacer que la distancia aumente cuadráticamente.
 Ejemplo: si hay muchas calles cerradas en la tarde, el camino se alarga más rápido con el tiempo.


−10x: Esto resta parte de esa distancia, tal vez porque al principio del día hay menos tráfico o algunos caminos están más directos.


+50: Esta es la distancia base de la ruta (en kilómetros, por ejemplo), si no hay desvíos ni tráfico.

En resumen: la distancia total depende de la hora del día. Algunas horas hacen la ruta más larga o más corta.
Esta función nos da la velocidad promedio, también en función de la hora.
-2x: Cada hora que pasa, la velocidad disminuye 2 km/h. Esto modela el aumento del tráfico durante el día.


+30: A la hora 0 (medianoche), la velocidad máxima es 30 km/h.
En resumen: mientras más tarde sales, más tráfico hay, y menos velocidad puedes mantener.




```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros para dos rutas: A (corta pero tráfico) y B (larga pero fluida)
# Ruta A: Distancia fija pero velocidad baja (por tráfico)
# Ruta B: Distancia más larga, velocidad más alta

# Definimos un rango de velocidades (en km/h)
v = np.linspace(10, 100, 500)

# Supongamos:
# Ruta A: d_A = 10 km, velocidad variable
# Ruta B: d_B = 20 km, velocidad mayor

# Tiempos en función de la velocidad
T_A = 10 / v  # Tiempo en horas
T_B = 20 / (v + 20)  # Simula que Ruta B tiene mejor velocidad media

# Gráfica comparativa
plt.figure(figsize=(10, 6))
plt.plot(v, T_A, label="Ruta A (corta, tráfico)", color="blue")
plt.plot(v, T_B, label="Ruta B (larga, fluida)", color="green")
plt.title("Comparación de tiempo de entrega según velocidad")
plt.xlabel("Velocidad promedio (km/h)")
plt.ylabel("Tiempo estimado (horas)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
